## 2. Написать DAG Airflow для ML-системы для расчета складских запасов сетевого магазина

1) Выберите архитектуру ML-конвейера (из архитектур в [задании 1](#cell-3f512028)).

2) Вспомните раздел Continuous Training Loop из конспекта этого модуля и напишите DAG Airflow для демонстрации концепции непрерывного переобучения (Continuous Training, CT).

3) ML-система должна **автоматически** (без участия человека) принимать решение о начале обучения, если:
* суммарно по всем кассам прошло более 10 млн чеков, и они загружены на S3;
* произошло незапланированное снижение метрики *accuracy < 0.85* из-за массового спроса на клон игрушки Лабубу;
* срок действия модели истекает через 1 час.


*Ожидаемый артефакт: код в [ячейке](#scrollTo=W0FaHkiQsW5J)*


```python
%%writefile inventory_retrain_dag.py

from datetime import datetime, timedelta

from airflow import DAG
from airflow.operators.python import BranchPythonOperator, PythonOperator


ACCURACY_THRESHOLD = 0.85
RECEIPTS_THRESHOLD = 10_000_000
MODEL_EXPIRATION_HOURS = 1


def wait_for_new_inventory_batch(**context):
    """
    In the full project this step checks MinIO/S3 bucket
    for a new daily inventory batch.
    """
    # Demo values. In the repository these values are read from metadata.json.
    context["ti"].xcom_push(key="batch_date", value="2026-06-13")
    context["ti"].xcom_push(key="total_receipts", value=12_500_000)
    context["ti"].xcom_push(key="batch_event", value="shock")
    return True


def load_inventory_data(**context):
    """
    Loads the new batch from S3-compatible storage.
    In the project MinIO is used as local S3 replacement.
    """
    print("Loading inventory batch from S3/MinIO")


def validate_inventory_data(**context):
    """
    Validates schema, row count, missing values and invalid ranges.
    """
    print("Inventory batch validation passed")


def evaluate_current_model_on_batch(**context):
    """
    Evaluates current production model on the new batch.
    Labubu-like demand shock may cause accuracy degradation.
    """
    # Demo value below threshold to show CT trigger.
    current_accuracy = 0.81
    context["ti"].xcom_push(key="current_accuracy", value=current_accuracy)
    return current_accuracy


def check_ct_conditions(**context):
    """
    Branching decision for Continuous Training.

    Retraining starts automatically if at least one risk condition is true:
    1. more than 10M receipts are available in S3/MinIO;
    2. current model accuracy dropped below 0.85;
    3. model expires in less than 1 hour.
    """
    ti = context["ti"]

    total_receipts = ti.xcom_pull(
        task_ids="wait_for_new_inventory_batch",
        key="total_receipts",
    )
    current_accuracy = ti.xcom_pull(
        task_ids="evaluate_current_model_on_batch",
        key="current_accuracy",
    )

    model_expires_at = datetime.utcnow() + timedelta(minutes=45)
    expires_soon = model_expires_at - datetime.utcnow() <= timedelta(
        hours=MODEL_EXPIRATION_HOURS
    )

    enough_receipts = total_receipts >= RECEIPTS_THRESHOLD
    quality_drop = current_accuracy < ACCURACY_THRESHOLD

    reasons = []
    if enough_receipts:
        reasons.append("receipts_over_10m")
    if quality_drop:
        reasons.append("accuracy_below_0_85")
    if expires_soon:
        reasons.append("model_expires_within_1h")

    print(
        "CT decision:",
        {
            "total_receipts": total_receipts,
            "current_accuracy": current_accuracy,
            "enough_receipts": enough_receipts,
            "quality_drop": quality_drop,
            "expires_soon": expires_soon,
            "reasons": reasons,
        },
    )

    if enough_receipts or quality_drop or expires_soon:
        return "train_model"

    return "skip_retraining"


def train_model(**context):
    print("Training new candidate model")


def evaluate_candidate_model(**context):
    candidate_accuracy = 0.88
    baseline_accuracy = 0.86

    context["ti"].xcom_push(key="candidate_accuracy", value=candidate_accuracy)
    context["ti"].xcom_push(key="baseline_accuracy", value=baseline_accuracy)

    print(
        {
            "candidate_accuracy": candidate_accuracy,
            "baseline_accuracy": baseline_accuracy,
        }
    )


def compare_with_baseline(**context):
    """
    Deployment decision is separated from retraining.
    Candidate model is registered only if it passes quality gate
    and is not worse than baseline.
    """
    ti = context["ti"]

    candidate_accuracy = ti.xcom_pull(
        task_ids="evaluate_candidate_model",
        key="candidate_accuracy",
    )
    baseline_accuracy = ti.xcom_pull(
        task_ids="evaluate_candidate_model",
        key="baseline_accuracy",
    )

    if candidate_accuracy >= ACCURACY_THRESHOLD and candidate_accuracy >= baseline_accuracy:
        return "register_model"

    return "skip_deploy"


def register_model(**context):
    print("Registering candidate model in model registry")


def skip_retraining(**context):
    print("Retraining skipped: CT conditions were not met")


def skip_deploy(**context):
    print("Deployment skipped: candidate model is unsafe")


with DAG(
    dag_id="inventory_retrain_pipeline",
    description="Continuous Training DAG for inventory stock forecasting",
    start_date=datetime(2026, 1, 1),
    schedule_interval="@daily",
    catchup=False,
    tags=["hw9", "continuous-training", "inventory"],
) as dag:

    t_wait_batch = PythonOperator(
        task_id="wait_for_new_inventory_batch",
        python_callable=wait_for_new_inventory_batch,
    )

    t_load = PythonOperator(
        task_id="load_inventory_data",
        python_callable=load_inventory_data,
    )

    t_validate = PythonOperator(
        task_id="validate_inventory_data",
        python_callable=validate_inventory_data,
    )

    t_eval_current = PythonOperator(
        task_id="evaluate_current_model_on_batch",
        python_callable=evaluate_current_model_on_batch,
    )

    t_check_ct = BranchPythonOperator(
        task_id="check_ct_conditions",
        python_callable=check_ct_conditions,
    )

    t_train = PythonOperator(
        task_id="train_model",
        python_callable=train_model,
    )

    t_eval_candidate = PythonOperator(
        task_id="evaluate_candidate_model",
        python_callable=evaluate_candidate_model,
    )

    t_compare = BranchPythonOperator(
        task_id="compare_with_baseline",
        python_callable=compare_with_baseline,
    )

    t_register = PythonOperator(
        task_id="register_model",
        python_callable=register_model,
    )

    t_skip_retraining = PythonOperator(
        task_id="skip_retraining",
        python_callable=skip_retraining,
    )

    t_skip_deploy = PythonOperator(
        task_id="skip_deploy",
        python_callable=skip_deploy,
    )

    (
        t_wait_batch
        >> t_load
        >> t_validate
        >> t_eval_current
        >> t_check_ct
    )

    t_check_ct >> t_train >> t_eval_candidate >> t_compare
    t_check_ct >> t_skip_retraining

    t_compare >> t_register
    t_compare >> t_skip_deploy
```

Рабочая полная версия DAG находится в репозитории: `dags/inventory_retrain_dag.py`.

В ячейке приведён основной фрагмент DAG, демонстрирующий Continuous Training Loop и автоматическое принятие решения о retraining по трём условиям задания: объём чеков более 10 млн, падение accuracy ниже 0.85 и истечение срока действия модели через 1 час.

# ADR 0001: Использовать MDD для решения по latency degradation

## Статус

Accepted.

## Контекст

В ML-системе прогнозирования складских остатков важна не только точность модели, но и системная задержка обработки/получения прогноза. Если latency становится слишком высокой, система перестаёт быть пригодной для регулярного использования в операционном процессе.

Для проверки был проведён MDD-эксперимент: сравнивались reference latency и new latency. В качестве SLI использовалась p95 latency. Для статистической проверки применялся Mann–Whitney U test.

## Результат

Новая версия показала значительный рост latency. p-value оказался ниже 0.05, а p95 latency превысила SLO 300 ms. На графике `reports/mdd_latency_distribution.png` видно, что распределение новой версии смещено вправо и содержит значения выше critical threshold 1000 ms.

## Решение

Не принимать новую версию как безопасную без архитектурных улучшений. Требуется снизить latency за счёт кеширования, предрасчёта признаков, оптимизации pipeline или разделения тяжёлых batch-операций и online/inference-части.

## Последствия

Решение принято не “на глаз”, а на основе метрики, статистического теста, графика и SLO. Это соответствует подходу Metrics Driven Development.
